```python
entries = len(df)
iot = range(8,19)
defns = range(34, 44)
for col in iot:
    for row in range(entries):
        if(df.iat[row, col] != 0):
            #df.iloc[row][column] = 1
            df.iat[row, col] = 1
for col in defns:
    for row in range(entries):
        if(df.iat[row, col] != 0):
            df.iat[row, col] = 1
```

```python
cleandf = pd.read_csv("cleanData.csv", encoding='cp1252', low_memory=False)
cleandf.head()
```

```python
defns = range(35,44)
totals = list()
countries = list()
for row in range(0, len(cleandf)): #raplace max with len(cleandf)
    defTot = 0
    for defs in defns:
        defTot += int(cleandf.iat[row, defs])
        #print("Row: " + str(row) + " Column: " + str(defs) + " Val: " + str(cleandf.iat[row, defs]))
    countries.append(cleandf.iat[row, 6])
    totals.append(defTot)

defsdf = pd.DataFrame({"country":countries, "total definitions":totals})
defsdf.head()
```

```python
defsdf = defsdf.sort_values(by="country")
defsdf.head()
```

```python
sections = list()
for i in range(0, len(defsdf)):
    if(defsdf.iat[i,0] != defsdf.iat[i-1,0]):
        print("Country change! From " + str(defsdf.iat[i-1,0]) + " to " + str(defsdf.iat[i,0]) + ".   Row#" + str(i))
        sections.append(i)
```

```python
print(sections)
avgs = list()
countries = list()
for i in range(1, len(sections)):
    total = 0
    for j in range(sections[i-1], sections[i]-1):
        total += defsdf.iat[j, 1]
    avg = total / (sections[i] - sections[i-1])
    avgs.append(avg)
    countries.append(defsdf.iat[sections[i], 0])
```

```python
avgdf = pd.DataFrame({"country":countries, "avg":avgs})
avgdf.head()
```
